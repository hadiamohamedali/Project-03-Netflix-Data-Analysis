```python
data['Show_id'] = data['Show_id'].apply(lambda x : x.split("s")[1])
```

```python
data["Show_id"] = data["Show_id"].astype(int)
data["Show_id"]
```

```python
data['Date_added'] = pd.to_datetime(data['Date_added'], errors='coerce')
```

```python
type(data['Date_added'])
```

```python
data['Date_added'].isna().sum()
```

```python
# Replace empty values ​​with the first day of the data
data['Date_added'].fillna(pd.Timestamp('2015-01-01'), inplace=True)
```

```python
data.columns
```

```python
# 1️⃣ Type (Movie / TV Show)
plt.figure(figsize = (6 , 5))
sns.countplot(x = 'Type' , data = data , palette = 'Set2')
plt.title('Distribution of Content Type')
plt.xlabel('Type')
plt.ylabel('Number of Titles')
plt.legend()
plt.show()
```

```python
# Piechart
data['Type'].value_counts().plot.pie(autopct='%1.1f%%', figsize=(5,5), colors=['#66b3ff','#ff9999'])
plt.ylabel('')
plt.title('Proportion of Movies vs TV Shows')
plt.legend()
plt.show()
```

```python
# 2️⃣ Rating
plt.figure(figsize = (10 , 5))
sns.countplot(x='Rating', data=data, order=data['Rating'].value_counts().index, palette='viridis')
plt.xticks(rotation = 45)
plt.title('Distribution of Ratings')
plt.xlabel('Rating')
plt.ylabel('Number of Titles')
plt.legend()
```

```python
# 3️⃣ Country
top_countries = data['Country'].value_counts().head(10)
plt.figure(figsize = (10 , 6))
sns.barplot(x = top_countries.values , y = top_countries.index , palette = 'magma')
plt.title('Top 10 Countries by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Country')
plt.legend()
plt.show()
```

```python
# 4️⃣ Genres
data['Genres'] = data['Listed_in']
genre_series = data['Genres'].str.split(', ').explode()
top_genres = genre_series.value_counts().head(10)
plt.figure(figsize = (10 , 6))
sns.barplot(x = top_genres.values , y = top_genres.index , palette = 'coolwarm')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Number of Titles')
plt.ylabel('Genre')
plt.legend()
plt.show()
```

```python
# 5️⃣ Director
top_directors = data['Director'].value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_directors.values, y=top_directors.index, palette='viridis')
plt.title('Top 10 Directors by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Director')
plt.show()
```

```python
# 6️⃣ Cast
cast_series = data['Cast'].str.split(', ').explode()
top_cast = cast_series.value_counts().head(10)
plt.figure(figsize=(10,6))
sns.barplot(x=top_cast.values, y=top_cast.index, palette='magma')
plt.title('Top 10 Actors/Actresses by Number of Titles')
plt.xlabel('Number of Titles')
plt.ylabel('Cast')
plt.show()
```

```python
# 7️⃣ Date Added
plt.figure(figsize=(12,6))
sns.histplot(data['Date_added'].dt.year, bins=10, kde=False, color='skyblue')
plt.title('Number of Titles Added per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()

plt.figure(figsize=(12,6))
sns.histplot(data['Date_added'].dt.month, bins=12, kde=False, color='salmon')
plt.title('Number of Titles Added per Month')
plt.xlabel('Month')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# 8️⃣ Release Year
plt.figure(figsize=(12,6))
sns.histplot(data['Release_year'], bins=20, kde=True, color='lightgreen')
plt.title('Distribution of Release Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.show()
```

```python
# 9️⃣ Duration
# Movies (minutes)
plt.figure(figsize=(10,5))
sns.histplot(data[data['Type']=='Movie']['Duration'], bins=30, kde=True, color='orange')
plt.title('Movie Duration Distribution (Minutes)')
plt.xlabel('Duration (Minutes)')
plt.ylabel('Number of Titles')
plt.show()

# TV Shows (Seasons)
plt.figure(figsize=(10,5))
sns.histplot(data[data['Type']=='TV Show']['Duration'], bins=10, kde=False, color='purple')
plt.title('TV Show Duration Distribution (Seasons)')
plt.xlabel('Number of Seasons')
plt.ylabel('Number of Titles')
plt.show()
```

```python
plt.figure(figsize=(12,6))
cross = pd.crosstab(data['Rating'], data['Type'])
sns.heatmap(cross, annot=True, fmt='d', cmap='viridis')
plt.title('Heatmap of Ratings vs Content Type')
plt.show()
```

```python
plt.figure(figsize=(12,6))
genre_year = data.groupby(['Date_added', 'Type']).size().unstack()
genre_year.plot(kind='line', marker='o')
plt.title('Titles Added per Year by Type')
plt.xlabel('Year Added')
plt.ylabel('Number of Titles')
plt.show()
```
